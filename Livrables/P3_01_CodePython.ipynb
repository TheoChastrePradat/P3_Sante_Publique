# Creation table_pivot bilan alimentaire

```python
# Import des librairies
import numpy as np
import pandas as pd
import seaborn as sns


# Import des données des bilans alimentaires
veg = pd.read_csv("vegetal.csv")
ani = pd.read_csv("animal.csv")

# Ajout de la variable origin
ani["origin"] = "animal"
veg["origin"] = "vegetal"

# On regroupe veg et ani en un unique dataframe, via une union
temp = ani.append(veg)

# Suppression de ani et veg
del ani, veg

# Transformation de temp en table pivot
data = temp.pivot_table(index=["Pays","Produit","origin"],
                        columns = ["Élément"], values=["Valeur"], aggfunc=sum)

data.columns = ['Aliments pour animaux', 'Autres Utilisations', 'Disponibilité alimentaire (Kcal/personne/jour)', 
                'Disponibilité alimentaire en quantité (kg/personne/an)', 
                'Disponibilité de matière grasse en quantité (g/personne/jour)', 
                'Disponibilité de protéines en quantité (g/personne/jour)', 'Disponibilité intérieure', 
                'Exportations - Quantité', 'Importations - Quantité', 'Nourriture', 'Pertes', 'Production', 
                'Semences', 'Traitement', 'Variation de stock']

data = data.reset_index()

colonnes_mult = ['Aliments pour animaux', 'Autres Utilisations', 'Disponibilité intérieure', 'Exportations - Quantité', 
                 'Importations - Quantité', 'Nourriture', 'Pertes', 'Production', 'Semences', 'Traitement', 'Variation de stock']
for elt in colonnes_mult:
    data[elt] *= 1000000

data.head()
```

### Remplacement des NaN par une valeur nul (0)

```python
data = data.fillna(value={"Aliments pour animaux":0,
                          "Autres Utilisations":0, 
                          "Disponibilité alimentaire (Kcal/personne/jour)":0, 
                          "Disponibilité alimentaire en quantité (kg/personne/an)":0,
                          "Disponibilité de matière grasse en quantité (g/personne/jour)":0, 
                          "Disponibilité de protéines en quantité (g/personne/jour)":0, 
                          "Exportations - Quantité":0, 
                          "Importations - Quantité":0, 
                          "Pertes":0, 
                          "Production":0, 
                          "Semences":0, 
                          "Traitement":0, 
                          "Variation de stock":0})
```

## Merge du dataFrame data et bpopu_2013_df

## Reorganisation des colonnes

```python
merge = merge[['Pays', 'Population', 'Produit', 'Nourriture', 'origin', 'Disponibilité alimentaire (Kcal/personne/jour)',
               'Disponibilité de protéines en quantité (g/personne/jour)']].copy()
```

## Ajout de deux nouvelles colonnes plus les valeurs attendues

```python
merge['Dispo Alim totale(Kcal)'] = merge['Population'] * merge['Disponibilité alimentaire (Kcal/personne/jour)'] * 365
merge.head()
```

```python
merge['Dispo Proteines Totale (g)'] = merge['Population'] * merge['Disponibilité de protéines en quantité (g/personne/jour)'] * 365
merge.head()
```

## Changement de nom du dataFrame en "tableau"

```python
tableau = merge.copy()
tableau.head()
```

# Q4

## Ajout de la colonne Ratio avec les valeurs attendues

```python
tableau['Ratio(Kcal/Kg)'] = tableau['Dispo Alim totale(Kcal)'] / tableau['Nourriture']
tableau['Ratio(g prot/Kg)'] = tableau['Dispo Proteines Totale (g)'] / tableau['Nourriture']
tableau.head()
```

```python
tableau = tableau.replace([np.inf, -np.inf], np.nan)
tableau = tableau.replace(0, np.nan)
tableau = tableau.groupby('Produit').mean()
tableau = tableau.reset_index()
```

# Q5

```python
tableau = tableau.fillna(0)
tableau.head()
```

```python
tableau = tableau[['Produit', 'Ratio(Kcal/Kg)', 'Ratio(g prot/Kg)']]
tableau.head()
```
